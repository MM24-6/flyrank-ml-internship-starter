## 4. Data limits

*What can this data never tell you? Unbalanced history, GSC-only early rows, window overlaps.*

Data Limit

The dataset has an unbalanced history because different clients started using Google Search Console (GSC) and Google Analytics 4 (GA4) at different times. In the loaded sample, all rows have GSC data available, but none have GA4 data available. This shows that early records do not always contain complete analytics data, so availability flags should always be checked before analysis.

```python
print("Rows loaded:", len(df))

print("\nMissing GSC availability:")
print(df["gsc_data_available"].value_counts(dropna=False))

print("\nMissing GA4 availability:")
print(df["ga4_data_available"].value_counts(dropna=False))
```
