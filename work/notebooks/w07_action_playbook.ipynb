```python
from huggingface_hub import login
login()

from datasets import load_dataset
import itertools
import pandas as pd

ds = load_dataset(
    "FlyRank/internship-warehouse",
    "fact_content_daily_performance",
    streaming=True,
    split="train"
)

rows = list(itertools.islice(ds, 50000))
df = pd.DataFrame(rows)

print(df.shape)
df.head()
```

## 1. Ranked actions + reason codes

*The queue: what to do first, and why, in words a human trusts.*

## Ranked Actions

Action 1:
Review and optimize pages with high impressions but low CTR because they have the greatest opportunity to improve search traffic.

Reason Code:
LOW_CTR_HIGH_IMPRESSIONS

Priority:
High

Expected Value:
Higher click-through rate and better organic traffic.

Human Decision:
A content reviewer should verify titles, meta descriptions, and page relevance before making changes.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
df["ctr"] = (df["gsc_clicks"] / df["gsc_impressions"].replace(0, 1)) * 100

df["baseline_score"] = df["gsc_impressions"] - df["ctr"]

df["reason_code"] = "LOW_CTR_HIGH_IMPRESSIONS"
df["action"] = "Review and optimize page"

ranked_queue = df.sort_values("baseline_score", ascending=False)

print(
    ranked_queue[
        [
            "gsc_impressions",
            "gsc_clicks",
            "ctr",
            "baseline_score",
            "reason_code",
            "action",
        ]
    ].head(10)
)
```
